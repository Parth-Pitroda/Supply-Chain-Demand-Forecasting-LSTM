```python
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader

class DemandDataset(Dataset):
    def __init__(self, data, lookback=30, horizon=7):
        """
        data: NumPy array of shape (total_days, num_features)
              The first column must be the target variable (sales demand).
        """
        self.data = torch.tensor(data, dtype=torch.float32)
        self.lookback = lookback
        self.horizon = horizon

    def __len__(self):
        return len(self.data) - self.lookback - self.horizon + 1

    def __getitem__(self, idx):
        # Input sequence includes all features across the lookback window
        x_start = idx
        x_end = idx + self.lookback
        X = self.data[x_start:x_end, :]

        # Target sequence is just the future demand (column 0)
        y_start = x_end
        y_end = x_end + self.horizon
        Y = self.data[y_start:y_end, 0]

        return X, Y
```

```python
import torch.nn as nn

class DemandForecasterLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, horizon=7):
        super(DemandForecasterLSTM, self,).__init__()
        self.hidden_dim = hidden_dim
        self.horizon = horizon

        # Encoder: Processes the historical window
        self.encoder = nn.LSTM(input_size=input_dim,
                               hidden_size=hidden_dim,
                               batch_first=True)

        # Decoder Cell: Forecasts the future one step at a time
        # Input to decoder cell is the previous prediction (1 dim)
        self.decoder_cell = nn.LSTMCell(input_size=1, hidden_size=hidden_dim)

        # Output project layer: Hidden state mapping back to a single demand scalar
        self.regression_layer = nn.Linear(hidden_dim, 1)

    def forward(self, X):
        batch_size = X.size(0)

        # 1. Pass historical data through Encoder
        _, (hn, cn) = self.encoder(X)

        # Squeeze out the layer dimension for the LSTMCell compatibility
        h_t = hn[-1]
        c_t = cn[-1]

        # 2. Use the last historical real value as the first decoder input
        decoder_input = X[:, -1, 0].unsqueeze(1) # Shape: (batch_size, 1)

        outputs = []

        # 3. Autoregressive decoding loop over the prediction horizon
        for t in range(self.horizon):
            h_t, c_t = self.decoder_cell(decoder_input, (h_t, c_t))
            prediction = self.regression_layer(h_t) # Shape: (batch_size, 1)
            outputs.append(prediction)

            # Use current prediction as the next step's input
            decoder_input = prediction

        # Stack predictions along the time dimension
        return torch.cat(outputs, dim=1) # Shape: (batch_size, horizon)
```

```python
def evaluate_wmape(model, dataloader, device):
    model.eval()
    total_absolute_error = 0.0
    total_actual_demand = 0.0

    with torch.no_grad():
        for X_batch, Y_batch in dataloader:
            X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)

            predictions = model(X_batch)

            # Sum up absolute differences across the entire batch and horizon
            total_absolute_error += torch.sum(torch.abs(Y_batch - predictions)).item()
            total_actual_demand += torch.sum(torch.abs(Y_batch)).item()

    wmape = (total_absolute_error / (total_actual_demand + 1e-8)) * 100
    return wmape
```

```python
# Generate 200 days of random sales data
np.random.seed(42)
total_days = 200

# Feature 0: Sales (random numbers between 50 and 200)
fake_sales = np.random.randint(50, 200, size=(total_days, 1))

# Feature 1: Day of the week (0, 1, 2, 3, 4, 5, 6, 0, 1...)
fake_dow = (np.arange(total_days) % 7).reshape(-1, 1)

# Feature 2: Weekend indicator (1 if day is 5 or 6, otherwise 0)
fake_weekend = (fake_dow >= 5).astype(int)

# Combine all 3 columns into a single data matrix
fake_data = np.hstack([fake_sales, fake_dow, fake_weekend])

print("Data matrix shape:", fake_data.shape)
print("First 3 days of data (Sales, DayOfWeek, IsWeekend):\n", fake_data[:3])
```

```python
# 1. Define the data-slicing logic
class DemandDataset(Dataset):
    def __init__(self, data, lookback=30, horizon=7):
        self.data = torch.tensor(data, dtype=torch.float32)
        self.lookback = lookback
        self.horizon = horizon

    def __len__(self):
        return len(self.data) - self.lookback - self.horizon + 1

    def __getitem__(self, idx):
        # Grab 30 days of history (all 3 features)
        X = self.data[idx : idx + self.lookback, :]

        # Grab the next 7 days of future sales (only column 0: Sales)
        Y = self.data[idx + self.lookback : idx + self.lookback + self.horizon, 0]
        return X, Y

# 2. Put our fake data into this slicer
dataset = DemandDataset(fake_data, lookback=30, horizon=7)

# 3. Use a DataLoader to feed data to our model in groups (batches) of 4
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# 4. Let's look at exactly one batch to make sure it works!
for X_batch, Y_batch in dataloader:
    print("History Batch Shape (X):", X_batch.shape)
    print("Future Target Shape (Y):", Y_batch.shape)
    break
```

```python
class DemandForecasterLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, horizon=7):
        super(DemandForecasterLSTM, self).__init__()
        self.hidden_dim = hidden_dim
        self.horizon = horizon

        # 1. The Encoder LSTM: reads historical data
        self.encoder = nn.LSTM(input_size=input_dim,
                               hidden_size=hidden_dim,
                               batch_first=True)

        # 2. The Decoder Cell: predicts the future day-by-day
        self.decoder_cell = nn.LSTMCell(input_size=1, hidden_size=hidden_dim)

        # 3. The Output Layer: converts neural brainwaves into a sales number
        self.regression_layer = nn.Linear(hidden_dim, 1)

    def forward(self, X):
        batch_size = X.size(0)

        # Pass data through the encoder
        _, (hn, cn) = self.encoder(X)
        h_t = hn[-1]
        c_t = cn[-1]

        # Start predictions using the very last known sales day
        decoder_input = X[:, -1, 0].unsqueeze(1)

        outputs = []
        # Predict day 1, day 2... up to day 7
        for t in range(self.horizon):
            h_t, c_t = self.decoder_cell(decoder_input, (h_t, c_t))
            prediction = self.regression_layer(h_t)
            outputs.append(prediction)
            decoder_input = prediction # feed current guess back as next input

        return torch.cat(outputs, dim=1)

print("Model class defined successfully!")
```

```python
def calculate_wmape(predictions, actuals):
    # Calculate how far off we are in absolute numbers
    total_absolute_error = torch.sum(torch.abs(actuals - predictions))
    total_actual_demand = torch.sum(torch.abs(actuals))

    # Avoid dividing by zero if data is empty
    wmape = (total_absolute_error / (total_actual_demand + 1e-8)) * 100
    return wmape.item()

print("Evaluation metric function defined successfully!")
```

```python
# 1. Set up the hardware tool (Use GPU if available, else CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 2. Initialize our model (3 input features, 16 hidden memory blocks, 7-day forecast)
model = DemandForecasterLSTM(input_dim=3, hidden_dim=16, horizon=7).to(device)

# 3. Define the standard loss function (MSE) and the weight updater (Adam)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

print("Starting training loop...\n")

# 4. Run the training for 10 epochs
for epoch in range(1, 11):
    model.train()
    epoch_loss = 0.0
    epoch_wmape = 0.0
    batches = 0

    for X_batch, Y_batch in dataloader:
        # Move data to the T4 GPU
        X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)

        # Reset the gradients from the last step
        optimizer.zero_grad()

        # Make predictions
        predictions = model(X_batch)

        # Calculate loss
        loss = criterion(predictions, Y_batch)

        # Backpropagation (the math that figures out how to improve)
        loss.backward()
        optimizer.step()

        # Keep track of scores
        epoch_loss += loss.item()
        epoch_wmape += calculate_wmape(predictions, Y_batch)
        batches += 1

    avg_loss = epoch_loss / batches
    avg_wmape = epoch_wmape / batches
    print(f"Epoch {epoch:02d} | Avg Loss (MSE): {avg_loss:.4f} | Avg Error (WMAPE): {avg_wmape:.2f}%")
```

```python
import matplotlib.pyplot as plt

# 1. Set the model to evaluation mode (stops it from training/updating weights)
model.eval()

# 2. Grab exactly one batch (4 examples) from our data pipeline
X_batch, Y_batch = next(iter(dataloader))
X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)

# 3. Ask the AI to predict the future based on the history
with torch.no_grad(): # Tells PyTorch not to calculate gradients (saves memory)
    predictions = model(X_batch)

# 4. Pull out the very first example in our batch and move it back to the CPU for plotting
# X_batch shape is (4, 30, 3). We want example 0, all 30 days, feature 0 (sales)
history = X_batch[0, :, 0].cpu().numpy()
actual_future = Y_batch[0, :].cpu().numpy()
predicted_future = predictions[0, :].cpu().numpy()

# 5. Create the timeline numbers for the bottom of the graph (the X-axis)
days_history = np.arange(1, 31)
days_future = np.arange(31, 38)

# 6. Draw the graph
plt.figure(figsize=(12, 6))
plt.plot(days_history, history, label="30-Day History (Actual Sales)", color="blue", marker="o")
plt.plot(days_future, actual_future, label="7-Day Future (Actual Sales)", color="green", marker="o")
plt.plot(days_future, predicted_future, label="7-Day Future (AI Prediction)", color="red", linestyle="--", marker="X", markersize=8)

# 7. Add labels and make it look professional
plt.title("Supply Chain Demand Forecasting: Actual vs. AI Prediction", fontsize=14, fontweight="bold")
plt.xlabel("Timeline (Days)", fontsize=12)
plt.ylabel("Sales Volume (Units)", fontsize=12)
plt.legend()
plt.grid(True, linestyle=":", alpha=0.7)
plt.show()
```

```python
# 1. Filter for just Store 1, Item 1 to start
single_item_data = real_data[(real_data['store'] == 1) & (real_data['item'] == 1)].copy()

# 2. Convert the 'date' text into a time format the computer understands
single_item_data['date'] = pd.to_datetime(single_item_data['date'])
single_item_data = single_item_data.sort_values('date') # Ensure chronological order

# 3. Create our temporal features (0=Monday, 6=Sunday)
single_item_data['day_of_week'] = single_item_data['date'].dt.dayofweek
single_item_data['is_weekend'] = (single_item_data['day_of_week'] >= 5).astype(int)

# 4. Isolate the exact 3 columns our AI brain expects: [Sales, DayOfWeek, IsWeekend]
model_ready_data = single_item_data[['sales', 'day_of_week', 'is_weekend']].values

print(f"Successfully filtered! New data shape: {model_ready_data.shape}")
print("First 3 days (Sales, DayOfWeek, IsWeekend):\n", model_ready_data[:3])
```

```python
# 1. Feed the REAL data into our sliding window slicer
real_dataset = DemandDataset(model_ready_data, lookback=30, horizon=7)

# We can use a larger batch size (16) because we have lots of real data now
real_dataloader = DataLoader(real_dataset, batch_size=16, shuffle=True)

# 2. Give the AI a fresh brain with slightly more memory (hidden_dim=32)
real_model = DemandForecasterLSTM(input_dim=3, hidden_dim=32, horizon=7).to(device)

# Define our optimizer
optimizer = torch.optim.Adam(real_model.parameters(), lr=0.005)

print("Starting training on REAL Store 1 data...\n")

# 3. Train for 15 epochs
for epoch in range(1, 16):
    real_model.train()
    epoch_wmape = 0.0
    batches = 0

    for X_batch, Y_batch in real_dataloader:
        X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)

        optimizer.zero_grad()
        predictions = real_model(X_batch)
        loss = criterion(predictions, Y_batch)
        loss.backward()
        optimizer.step()

        epoch_wmape += calculate_wmape(predictions, Y_batch)
        batches += 1

    avg_wmape = epoch_wmape / batches
    print(f"Epoch {epoch:02d} | Avg Error (WMAPE): {avg_wmape:.2f}%")
```
